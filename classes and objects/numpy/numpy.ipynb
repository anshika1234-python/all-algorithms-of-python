```python
img3=img[::-1,::-1,:]   # reverse the rows and columns
plt.imshow(img3) 
```

```python
img3=img[::-1,:,:]   # rotate the image right 180 degrees
plt.imshow(img3) 
```

```python
#Problem: Given two 2D arrays, stack them vertically.

import numpy as np
a = np.arange(10).reshape(2, 5)
b = np.repeat(2, 10).reshape(2, 5)
vstacked = np.vstack([a, b])
print(vstacked)
```

```python
# Problem: Given two 2D arrays, stack them horizontally.

import numpy as np
a = np.arange(10).reshape(2, 5)
b = np.repeat(2, 10).reshape(2, 5)
hstacked = np.hstack([a, b])
print(hstacked)
```

```python
# Create a sequence that repeats and tiles a given array.

import numpy as np
a = np.array([1, 2, 3])
result = np.r_[np.repeat(a, 3), np.tile(a, 3)]
print(result)
```

```python
# Remove elements from one array that are present in another array.

import numpy as np
a = np.array([1, 2, 3, 4, 5])
b = np.array([5, 6, 7, 8, 9])
result = np.setdiff1d(a, b)
print(result)
```
